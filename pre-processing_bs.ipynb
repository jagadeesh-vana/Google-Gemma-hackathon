```python
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd

def extract_telugu_text_from_section(url):

  # Send a GET request to the webpage
    response = requests.get(url)
    response.encoding = response.apparent_encoding  # Handle proper encoding for Telugu text

    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        section = soup.find("div", class_="split")
        h2_tag = soup.find('h1', class_="entry-title")

       # Extract story
        if section:
            text_content = section.get_text()
            story = ''.join(char for char in text_content if '\u0C00' <= char <= '\u0C7F' or char in [' ', '\n', '.', ','])
        else:
            story = f"Could not find the specified story section in: {url}"

        # Extract title
        if h2_tag:
            telugu_text = h2_tag.get_text()
            title = telugu_text.split('-')[1]
        else:
            title = f"Could not find the specified title section in: {url}"
        return title, story
    else:
        print(f"Failed to fetch the webpage. Status code: {response.status_code} : url{url}")
        return None, None
```

```python
sts = []
for i in range(100):
  url = "https://tsstudies.blogspot.com/2020/04/chandamama-kathalu_" + str(i) + ".html"
  sts.append(extract_telugu_text_from_section(url))
stories = pd.DataFrame(sts, columns=['title', 'story'])
stories
```
